```python
'''
Analyze class imbalance in chest MNIST challenge19

'''
import time
start_time=time.time()
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
# Load the .npz file
data = np.load('all_bucket/marketplace-c3kdc8vj/MedMNIST/120/artifactFiles/chestmnist.npz')

# Extract train images and labels
train_images = data['train_images']
train_labels_onehot = data['train_labels']

print(f"Total training images: {len(train_images)}")
print(f"Image shape: {train_images[0].shape}")
print(f"Label shape: {train_labels_onehot.shape}")

# Convert one-hot encoded labels to class indices
train_labels = np.argmax(train_labels_onehot, axis=1)

# Count the number of samples in each class
label_counts = Counter(train_labels)

# Define the label mapping
label_mapping = ['atelectasis', 'cardiomegaly', 'effusion', 'infiltration', 'mass', 
                 'nodule', 'pneumonia', 'pneumothorax', 'consolidation', 'edema', 
                 'emphysema', 'fibrosis', 'pleural', 'hernia']

# Create a list of counts for each class
counts = [label_counts[i] for i in range(len(label_mapping))]

# Create a bar plot
plt.figure(figsize=(14, 6))
plt.bar(label_mapping, counts)
plt.title('Number of Samples per Class in Training Set')
plt.xlabel('Class Label')
plt.ylabel('Number of Samples')

# Add value labels on top of each bar
for i, count in enumerate(counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
end_time=time.time()
print("total_time:" , end_time-start_time)
```
